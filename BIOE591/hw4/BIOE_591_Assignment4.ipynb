If evapotranspiration from Earth's land area were to diminish by 20% uniformly over the land area, as might result from widespread removal of vegetation, what changes would occur in the globally averaged precipitation on the land surface and in the globally averaged runoff from the land to the sea?

```python
P_L = 108e3# rate of precipitation km^3/yr
P_S = 410e3 # rate of precipitation on the sea km^3/yr
R = 46e3 # rate of runoff from the land to the sea km^3/yr

#make the assumptions that E_LL = K * E_LS

K = 3 #the proportion of land to sea evapotranspiration compared to land to land evapotranspiration
E_LS = (P_L - R) / (K + 1) # rate of evapotranspiration from the land that falls as precipitation on the sea
E_LL = 3 * E_LS # rate of evapotranspiration from the land that falls as precipitation on the land
E_SS = P_S - E_LS # rate of evapotranspiration from the sea that falls as precipitation on the sea
E_SL = P_L - E_LL # rate of evapotranspiration from the sea that falls as precipitation on the land
```

Now assume there is a 20% loss in the E_LL and E_LS values

```python
reduction = .2
prop = 1- reduction
Ep_LL = E_LL * prop
Ep_LS = E_LS * prop
```

```python
Rp = E_SL - Ep_LS
Pp_L = Ep_LL + E_SL
percent_R = (Rp-R)/R
percent_PL = (P_L-Pp_L)/P_L
print('The change in total runoff with a %.1f%% reduction in evapotraspiration is %.1f km^3 yr-1 or %.1f%%' %(reduction*100, Rp-R, percent_R*100))
print('The change in land precipitaion with a %.1f%% reduction in evapotraspiration is %.1f km^3 yr-1 or %.1f%%' %(reduction*100, P_L-Pp_L, percent_PL*100))
```

Deforestation of mountain slopes has been proposed as a way to provide more runoff water for irrigation in the western United States. The idea is that more snowmelt will flow into reservoirs while less will be transpired away by trees. Is this a sensible idea?

Considering the examples listed from [*The Water Resources Bulletin*, Volume 19, June, 1983](http://coweeta.uga.edu/publications/1025.pdf), this general principle of deforestation for more runoff water does have empirical evidence:
>*Water Yield Augmentation from Hardwoods
We know from scientific studies begun nearly 50 years ago
that interruption of transpiration saves water, which is eventually
released to ground water or streams. Hibbert (1967)
summarized worldwide catchment studies and observed that
deforestation increased and afforestation decreased water
yield. He concluded that the response was highly variable
and, for the most part, unpredictable. This might be expected
when examining data from the great range of climates, vegetations,
and geomorphic conditions which exist worldwide. Lull
and Reinliart (1967) reviewed available information from
forested catchment studies in the Northeastern United States
looking for effects of partial or complete forest removal on
yield response. They concluded that first year water yield
increases were not highly sensitive to precipitation amount,
that partial cuttings were not as efficient for augmenting water
yield as were complete cuttings, and that water yield from
well-stocked northeastern forests could be increased by from
4 to 12 inches the first year after complete cutting.*

So, it is **possible** to remove trees to increase runoff, however the question is regarding **sensibility**. Do I feel like it is a reasonable ecosystem service loss of forest versus increases streamflow yield? My initial reaction is, no. Upon, further literature review regarding experiments at the Hubbard Brook experimental forest by [Hornbeck and Kochenderfer(2001)](http://www.fs.fed.us/ne/newtown_square/publications/other_publishers/ne_2001_hornbeck001p.pdf), its seems that it is possible to increase runoff through deforestation, but you run into a number of concerns:

1. Increased timing of spring runoff. Peak flow could arrive earlier in the season, resulting in low streamflows by the late summer, when water yields may be essential for certain wildlife species.

2. Nutrient cycling may change drastically, with potentially high increases in streamwater concentrations of nitrogen (as nitrate), hydrogen, and base cations. This nutrient increase may result in eutrophication to downstream lakes. 

3. Changes in forest type could occur after major deforestation within a region. Firstly, to maintain the increased water yield from deforesting a region, one must continually apply herbicide, so regrowth does not take back those water yield gains. This means a continual effort must be made to deforest a area every few years. Secondly, if regrowth is unmanaged, then the returning vegetation may be drastically different than what was previously there, impacting the habitat suitability for local wildlife. 

For these reasons, I personally feel that it is not sensible to risk the potential adverse effects from deforestation for a short gain in water yield.
